```python
%%writefile app.py
import streamlit as st
from transformers import pipeline
import torch

# UI Page Config
st.set_page_config(page_title="Smart AI Assistant", page_icon="🧠", layout="centered")
st.title("🧠 Smart AI Chatbot")
st.caption("🤖 Powered by Qwen-2.5 (Instruct) | Expert in GK, Math & AI/ML definitions")

# 1. Load the Advanced Instruct Model on GPU
@st.cache_resource
def load_smart_model():
    # pipeline automatically uses the GPU via device_map="auto"
    return pipeline(
        "text-generation",
        model="Qwen/Qwen2.5-1.5B-Instruct",
        torch_dtype=torch.float16,
        device_map="auto"
    )

pipe = load_smart_model()

# 2. System Prompt to define the Bot's personality and knowledge base
if "messages" not in st.session_state:
    st.session_state.messages = [
        {
            "role": "system",
            "content": "You are a highly intelligent AI assistant. You excel at general knowledge, precise mathematical calculations, and explaining complex AI/ML and engineering definitions in simple terms with examples."
        }
    ]

# 3. Display Chat History (Hide the hidden system prompt from the user)
for message in st.session_state.messages:
    if message["role"] != "system":
        with st.chat_message(message["role"]):
            st.markdown(message["content"])

# 4. Chat Input & Processing
if user_prompt := st.chat_input("Ask me a math problem, GK question, or AI definition..."):
    # Display user message
    with st.chat_message("user"):
        st.markdown(user_prompt)

    # Add to session history
    st.session_state.messages.append({"role": "user", "content": user_prompt})

    # Generate response
    with st.chat_message("assistant"):
        with st.spinner("Analyzing and calculating..."):
            # Pass the entire conversation context to the model
            outputs = pipe(
                st.session_state.messages,
                max_new_tokens=512,
                do_sample=True,
                temperature=0.7,
                top_p=0.9
            )

            # Extract only the latest response from the model's output structure
            bot_response = outputs[0]["generated_text"][-1]["content"]
            st.markdown(bot_response)

    # Add bot response to session history
    st.session_state.messages.append({"role": "assistant", "content": bot_response})
```
